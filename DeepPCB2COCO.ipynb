User Defined base functions Code for the above description

```python

def class_dict():
    class_dict={1:"open", 2:"short", 3:"mousebite", 4:"spur", 5:"copper", 6:"pin-hole"}
    return class_dict
def info_dict_obj(description = None, url = None, version = None, year = None, contributor = None, date_created = None):
    info_dict_object = {
            "description": description,
            "url": url,
            "version": version,
            "year": year,
            "contributor": contributor,
            "date_created": date_created
            }
    return(info_dict_object)
   
def licenses_dict_obj(url=None,Lid=None,name=None):
    
    licenses_dict_object = {
        "url": url,
        "id": Lid,
        "name": name
        }
    return(licenses_dict_object)
    '''else:
        return None'''   
def image_dict_obj(image_id=None,license=None,coco_url=None,flickr_url=None, height= None,width = None, filename=None, date_captured = None):
    image_dict_object = {
            "id" : image_id,
            "license": license,
            "coco_url": coco_url,
            "flickr_url": flickr_url,
            "height" : height,
            "width" : width,
            #['group20085,"20085","20085000_test.jpg"], 'group20085/20085_not/20085000.txt']

            "file_name": filename.split('/')[-1],
            "date_captured": date_captured,
            }
    return(image_dict_object)   
   
def categories_dict_obj(super_category_name = None, category = None, category_name = None):
    categories_dict_object = {}
    if super_category_name==None:
        super_category_name=category_name
        categories_dict_object={
                "supercategory": super_category_name,
                "id": category,
                "name": category_name
            }
    else:
        categories_dict_object={
                "supercategory": super_category_name,
                "id": category,
                "name": category_name
            }
    return(categories_dict_object)   
   
def annotation_dict_obj(annotation_id=None,category=None,iscrowd=0,segmentation=[],image_id=None,area=None,boxes=None):
    annotation_dict_object = {
                "id": annotation_id,
                "category_id": category,
                "iscrowd": iscrowd,
                "segmentation": segmentation,
                "image_id" : image_id,
                "area": area,
                "bbox": boxes
            }
    return(annotation_dict_object)   
   
def dict_of_COCO_format(info_obj,licenses_objs,image_objs,categories_objs,annotation_objs):
    COCO_Format_dict={}
    COCO_Format_dict["info"] = info_obj
    COCO_Format_dict["licenses"] = licenses_objs
    COCO_Format_dict["images"] = image_objs
    COCO_Format_dict["annotations"] = annotation_objs
    COCO_Format_dict["categories"] = categories_objs

    return COCO_Format_dict
                           
```

# **get_PCB_dict function**

(This is a **USER DEFINED FUNCTION** )

---

### **INPUT** : list (list created by reading the trainval.txt/test.txt)

---


 

#### **Used for** : CONVERTING the **LIST** created using **trainval.txt** & **test.txt** into a dictionary that is in **COCO FORMAT**


---




### **OUTPUT** : Dictionary (that is in COCO format)

```python
def get_PCB_dict(data_list):
    dataset_dicts = {}
    dataset_final_dicts={}
    
    
    #info
    year= 2022
    version= '1.0'
    description= 'DeepPCB-2-COCO-Format-2022'
    contributor= ''
    url= ''
    date_created= '2022/1/10'
    
    info_obj = info_dict_obj(description, url, version, year, contributor, date_created)

    
    # Adding each dictionary object to to the Dictionary dataset_dicts
    #dataset_dicts["info"] = info_obj
    #---------------------------------------------------------------------
    
    #licenses

    i=0
    url=''
    Lid=0
    name=''
    licenses_objs=[]
    licenses_obj=licenses_dict_obj(url,Lid,name)
    licenses_objs.append(licenses_obj)

    
    # Adding each dictionary object to to the Dictionary dataset_dicts
    #dataset_dicts["licenses"] = licenses_objs
    #---------------------------------------------------------------------

    categories_dict=class_dict()
    categories_objs = []
    super_category_name=None
    for key, value in categories_dict.items():
        #Categories
        category=key
        category_name=value
        categories_obj=categories_dict_obj(super_category_name,category,category_name)
        categories_objs.append(categories_obj)

    #---------------------------------------------------------------------
    
    #annotations
    annotation_objs = []
    annotation_id=0
    id_of_image=None
    segmentation=None
    area=None
    iscrowd=0

    #---------------------------------------------------------------------
    
    #images
    image_objs = []
    image_id=1000
    coco_url=''
    flickr_url=''
    
    date_captured = "2022"
    for i,path in enumerate(data_list):
        filename = '/content/drive/MyDrive/Datasets/PCBData/'+path[0]    
        height, width = cv2.imread(filename).shape[:2]

        image_obj = image_dict_obj(image_id,Lid,coco_url,flickr_url, height, width, filename, date_captured)
        image_objs.append(image_obj)

        
        # Adding each dictionary object to to the Dictionary dataset_dicts
        #dataset_dicts["images"] = image_objs
        
        #['group20085/20085/20085000_test.jpg', 'group20085/20085_not/20085000.txt']

        with open('/content/drive/MyDrive/Datasets/PCBData/'+path[1]) as t:
            lines = t.readlines()
            for line in lines:
                if line[-1]=="\n":
                  box = line[:-1].split(' ')
                # ["466","441","493","470","3",""]
                
                else:
                  box = line.split(' ')

                # ["466","441","493","470","3"]



                #[top left x position, top left y position, width, height]
                bbox_width =float(box[2])-float(box[0])
                bbox_height=float(box[3])-float(box[1])
                boxes = list(map(float,[box[0],box[1],str(bbox_width),str(bbox_height)]))
                
                category = int(box[4])
                

                #---------------------------------------------------------------------
                  
                segmentation=[]
                area=0.0
                #Annotations
                annotation_obj = annotation_dict_obj(annotation_id,category,iscrowd,segmentation,image_id,area,boxes)
                annotation_id+=1
                annotation_objs.append(annotation_obj)

            image_id+=1

                #---------------------------------------------------------------------

            
            # Adding each dictionary object to to the Dictionary dataset_dicts
            #dataset_dicts["categories"] = categories_objs
            #dataset_dicts["annotations"] = annotation_objs
        
    
    dataset_dicts=dict_of_COCO_format(info_obj,licenses_objs,image_objs,categories_objs,annotation_objs)
    
    #dataset_final_dicts = dataset_final_dicts + (dataset_dicts,)

    return dataset_dicts #dict of dicts
```

```python
import os
import json

import numpy as np
from numpy.core.fromnumeric import mean


def get_metadata_frm_json(json_data, label_names):
    
    categories = json_data["categories"]
    annotations=json_data["annotations"]

    label_counts_v2={}
    for annotation in annotations:
        if annotation["category_id"] in label_counts_v2:
            label_counts_v2[annotation["category_id"]]+=1
        else:
            label_counts_v2[annotation["category_id"]]=1
    print(label_counts_v2)

    
    
    unique_label_names = np.unique(np.array(label_names))
    label_counts = {}
    for label_name in unique_label_names:
        label_counts[label_name] = 0
        for category in categories:
            if category["name"] == label_name:
                label_counts[label_name] += 1
    print(unique_label_names)
    return unique_label_names, label_counts_v2


path = "/content/drive/MyDrive/JSONdump"
train_json_filename = path+"/"+"instances_train2014.json"
val_json_filename = path+"/"+"instances_test2014.json"
metadata_json_filename = path+"/"+"metadata.json"

# read jsons
with open(train_json_filename, "r") as f:
    train_json_data = json.load(f)

with open(val_json_filename, "r") as f:
    val_json_data = json.load(f)
with open(metadata_json_filename, "r") as f:
    metadata_json_data = json.load(f)

labels = ["open", "short", "mousebite", "spur", "copper", "pin-hole"]

# iterate and get the data
unique_label_names, label_counts = get_metadata_frm_json(train_json_data, labels)

metadata_json_data["labelDistribution"]["TRAIN"]["annotationCounts"] = label_counts
metadata_json_data["labelDistribution"]["TRAIN"]["labelNames"] = unique_label_names.tolist()
metadata_json_data["labelDistribution"]["TRAIN"]["imagesCount"] = len(train_json_data["images"])


unique_label_names, label_counts = get_metadata_frm_json(val_json_data, labels)

metadata_json_data["labelDistribution"]["VALIDATION"]["annotationCounts"] = label_counts
metadata_json_data["labelDistribution"]["VALIDATION"]["labelNames"] = unique_label_names.tolist()
metadata_json_data["labelDistribution"]["VALIDATION"]["imagesCount"] = len(val_json_data["images"])


with open(metadata_json_filename, "w") as file_pointer_train:
    json.dump(metadata_json_data, file_pointer_train,indent = 4)

s = 1
```
